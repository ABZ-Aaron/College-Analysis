Let's practice creating some Histograms to explore distributions.

```python
data_cols = ['Sample_size',
             'Median',
             'Employed',
             'Full_time',
             'ShareWomen',
             'Unemployment_rate',
             'Men',
             'Women']

fig = plt.figure(figsize = (15, 20))

for i in range(8):
    
    ax = fig.add_subplot(4, 2, i+1)
    ax.hist(recent_grads[data_cols[i]], bins = 15)
    
    value = data_cols[i].replace("_", " ").title()
    ax.set_title(value)
```

That's one way to do it. We can also do something like this...

```python
ax = recent_grads['Sample_size'].hist(bins = 15, figsize = (5, 5), range = (0, 2000))
ax.set_title("Sample Size")
plt.show()
```

or this...

```python
ax = recent_grads['Sample_size'].plot(kind = 'hist', title = "Sample Size", figsize = (5, 5))
plt.show()
```

From observing these distributions, some appaer close to a normal distribution (bell shape), while others are heavily skewed.


We can also generate a `scatter matrix plot` which combines both scatter plots and histograms. Let's have a look.

```python
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize = (10, 10))
plt.show()
```

We can also change the histogrames to kernal density estimations

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize = (15, 15), diagonal = 'kde')
plt.show()
```

Let's have a quick look at bar charts

We can also do this...

We may wnat to assign some labels...

```python
recent_grads[:10].plot.bar(y = "ShareWomen", x = "Major_category", title = "Share of Women", legend = False)
plt.show()
```

In reality, we would probably want to group this data and present it differently, like below...

```python
major_groupby = recent_grads.groupby(['Major_category'])['ShareWomen'].mean().sort_values()
ax = major_groupby.plot(kind = 'barh', rot = 0, title = 'Share of Women by Major Category', figsize = (10, 10))
ax.set_ylabel("Major Category")
plt.show()
```

Let's look closer at the number of women and men in each major category

```python
major_groupby = recent_grads.groupby(['Major_category'])['Women'].mean().sort_values()
ax = major_groupby.plot(kind = 'barh', rot = 0, title = 'No of Women by Major Category', figsize = (10, 10))
ax.set_ylabel("Major Category")
plt.show()
```

```python
major_groupby = recent_grads.groupby(['Major_category'])['Men'].mean().sort_values()
ax = major_groupby.plot(kind = 'barh', rot = 0, title = 'No of Men by Major Category', figsize = (10, 10))
ax.set_ylabel("Major Category")
```

We might be better place these on the same figure, for an easier comparison. Let's have a look.

```python
section = recent_grads[['Major_category', 'Men', 'Women']]
gender_category = section.groupby(['Major_category']).mean().sort_index().astype(int)
gender_category
```

```python
ax = gender_category.plot(kind = 'barh', figsize = (10, 10))
ax.set_title("Men & Women by Major Category")
ax.set_ylabel("Major Category")
ax.set_xlabel("Total")
plt.show()
```

```python
fig = plt.figure(figsize = (15, 5))

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.boxplot(x = recent_grads['Unemployment_rate'])
ax1.set_title('Unemployment Rate')
ax1.set_ylabel('Rate of Unemployment')

ax2.boxplot(x = recent_grads['Median'])
ax2.set_title('Median Salary')
ax2.set_ylabel('Salary (Median)')

plt.show()
```

Use a hexagonal bin plot to visualize the columns that had dense scatter plots from earlier in the project. Let's do it slighly different than previously (borrowed from someone else's project).

Let's also play around with the colours a bit.

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(20,20))
plt.subplots_adjust(hspace=.5)

recent_grads.plot(x = 'Sample_size', y = 'Median', kind = 'hexbin', ax = axes[0,0], gridsize=25, colormap = 'magma', rot=45, sharex = False)
recent_grads.plot(x = 'Sample_size', y = 'Unemployment_rate', kind = 'hexbin', ax = axes[0,1], gridsize=25, colormap = 'plasma', rot=45, sharex = False)
recent_grads.plot(x = 'Full_time', y = 'Median', kind = 'hexbin', ax = axes[1,0], gridsize=25, colormap = 'viridis', rot=45, sharex = False)
recent_grads.plot(x = 'ShareWomen', y = 'Unemployment_rate', kind = 'hexbin', ax = axes[1,1], gridsize=25, colormap = 'cividis', rot=45, sharex = False)
recent_grads.plot(x = 'Men', y = 'Median', kind = 'hexbin', ax = axes[2,0], gridsize=25, colormap = 'inferno', rot=45, sharex = False)
recent_grads.plot(x = 'Women', y = 'Median', kind = 'hexbin', ax = axes[2,1], gridsize=25, colormap = 'winter', rot=45, sharex = False)

plt.show()
```
